```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, auc
)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Split data into 70% train and 30% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Step 2: Cross-validation on training set
model = LogisticRegression(max_iter=1000)
cv = StratifiedKFold(n_splits=5)

cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
print("Cross-Validation Accuracy Scores (Train Set):", cv_scores)
print("Mean CV Accuracy:", np.mean(cv_scores))

# Step 3: Train final model on full training data
model.fit(X_train, y_train)

# Step 4: Evaluate on test set
y_pred_lr = model.predict(X_test)
y_prob_lr = model.predict_proba(X_test)[:, 1]

# Step 5: Metrics
accuracylr = accuracy_score(y_test, y_pred_lr)
precisionlr = precision_score(y_test, y_pred_lr)
recalllr = recall_score(y_test, y_pred_lr)
f1lr = f1_score(y_test, y_pred_lr)
results = {}
results["Logistic Regression"] = (accuracylr, precisionlr, recalllr, f1lr)

print("\nFinal Logistic Regression Results on Test Set")
print("Accuracy:", accuracylr)
print("Precision:", precisionlr)
print("Recall:", recalllr)
print("F1-Score:", f1lr)

# Step 6: Confusion Matrix
cm = confusion_matrix(y_test, y_pred_lr)
plt.figure()
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Logistic Regression - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Step 7: ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob_lr)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend()
plt.grid(True)
plt.show()
```
